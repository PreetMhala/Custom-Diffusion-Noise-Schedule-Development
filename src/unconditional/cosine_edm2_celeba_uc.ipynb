```python
# ============================================================
# ✅ PATCH: COSINE TRAINING NOISE SCHEDULE
# ============================================================
import torch, math
import training.training_loop as loop

def cosine_sigma(batch_size, device, sigma_min=0.002, sigma_max=80):
    """
    Cosine decay schedule between sigma_max → sigma_min.
    sigma(t) = sigma_min + (sigma_max - sigma_min) * (1 + cos(pi*t)) / 2
    """
    t = torch.rand(batch_size, 1, 1, 1, device=device)
    cos_term = (1 + torch.cos(math.pi * t)) / 2
    return sigma_min + (sigma_max - sigma_min) * cos_term


def patched_call(self, net, images, labels=None):
    batch = images.shape[0]

    # === COSINE NOISE ===
    sigma = cosine_sigma(batch, images.device)

    # === EDM2 weighting ===
    weight = (sigma**2 + self.sigma_data**2) / (sigma * self.sigma_data)**2

    # Add noise
    noise = torch.randn_like(images) * sigma

    # Forward
    denoised, logvar = net(images + noise, sigma, labels, return_logvar=True)

    # Loss
    loss = (weight / logvar.exp()) * ((denoised - images)**2) + logvar
    return loss


# Inject patch
loop.EDM2Loss.__call__ = patched_call
print("✅ Cosine training schedule injected successfully!")
```
